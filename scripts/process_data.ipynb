```
import pandas as pd
df=pd.DataFrame
df = pd.read_csv('data.csv')
```

```
Business = df[['PAYANN','PAYANN_S','EMP','EMP_S','YEAR']]
Business.rename(columns={'PAYANN':'annual_payroll','PAYANN_S':'payroll_error','EMP':'num_employees','EMP_S':'employees_error','YEAR':'year'}, inplace=True)
Business['id'] = Business.index
Business['location_id'] = Business.index
Business['industry_id'] = Business.index
Business['ethnicity_id'] = Business.index
Business['race_id'] = Business.index
Business['gender_id'] = Business.index
Business['firm_id'] = Business.index
Business['receipts_id'] = Business.index
Business = Business.drop(Business.index[0])
Business = Business[['id','location_id','industry_id','ethnicity_id','race_id','gender_id','firm_id','receipts_id','annual_payroll','payroll_error','num_employees','employees_error','year']]
Business.head(10)
```

```
Location = pd.DataFrame(data=(df.get('GEO_TTL')))
Location.rename(columns={'GEO_TTL':'location'}, inplace=True)
Location['id'] = Location.index
Location = Location.drop(Location.index[0])
Location = Location[['id','location']]
Location.head(10)
```

```
Industry = pd.DataFrame(data=(df.get('NAICS2012_TTL')))
Industry.rename(columns={'NAICS2012_TTL':'industry'}, inplace=True)
Industry['id'] = Industry.index
Industry = Industry.drop(Industry.index[0])
Industry = Industry[['id','industry']]
Industry.head(10)
```

```
Ethnicity = pd.DataFrame(data=(df.get('ETH_GROUP_TTL')))
Ethnicity.rename(columns={'ETH_GROUP_TTL':'ethnic_group'}, inplace=True)
Ethnicity['id'] = Ethnicity.index
Ethnicity = Ethnicity.drop(Ethnicity.index[0])
Ethnicity = Ethnicity[['id','ethnic_group']]
Ethnicity.head(10)
```

```
Race = pd.DataFrame(data=(df.get('RACE_GROUP_TTL')))
Race.rename(columns={'RACE_GROUP_TTL':'race_group'}, inplace=True)
Race['id'] = Race.index
Race = Race.drop(Race.index[0])
Race = Race[['id','race_group']]
Race.head(10)
```

```
Gender = pd.DataFrame(data=(df.get('SEX_TTL')))
Gender.rename(columns={'SEX_TTL':'sex'}, inplace=True)
Gender['id'] = Gender.index
Gender = Gender.drop(Gender.index[0])
Gender = Gender[['id','sex']]
Gender.head(10)
```

```
Firm = df[['FIRMALL','FIRMALL_S','FIRMPDEMP','FIRMPDEMP_S','FIRMNOPD','FIRMNOPD_S']]
Firm.rename(columns={'FIRMALL':'total_firms','FIRMALL_S':'total_error','FIRMPDEMP':'paid_firms','FIRMPDEMP_S':'paid_error','FIRMNOPD':'unpaid_firms','FIRMNOPD_S':'unpaid_error'}, inplace=True)
Firm['id'] = Firm.index
Firm = Firm.drop(Firm.index[0])
Firm = Firm[['id','total_firms','total_error','paid_firms','paid_error','unpaid_firms','unpaid_error']]
Firm.head(10)
```

```
Receipts = df[['RCPALL','RCPALL_S','RCPPDEMP','RCPPDEMP_S','RCPNOPD','RCPNOPD_S']]
Receipts.rename(columns={'RCPALL':'total_receipts','RCPALL_S':'total_error','RCPPDEMP':'paid_receipts','RCPPDEMP_S':'paid_error','RCPNOPD':'unpaid_receipts','RCPNOPD_S':'unpaid_error'}, inplace=True)
Receipts['id'] = Receipts.index
Receipts = Receipts.drop(Receipts.index[0])
Receipts = Receipts[['id','total_receipts','total_error','paid_receipts','paid_error','unpaid_receipts','unpaid_error']]
Receipts.head(10)
```

```
Business.to_csv('business.csv',index=False,header=True)
files.download('business.csv')
Industry.to_csv('industry.csv',index=False,header=True)
files.download('industry.csv')
Location.to_csv('location.csv',index=False,header=True)
files.download('location.csv')
Ethnicity.to_csv('ethnicity.csv',index=False,header=True)
files.download('ethnicity.csv')
Race.to_csv('race.csv',index=False,header=True)
files.download('race.csv')
Gender.to_csv('gender.csv',index=False,header=True)
files.download('gender.csv')
Firm.to_csv('firm.csv',index=False,header=True)
files.download('firm.csv')
Receipts.to_csv('receipts.csv',index=False,header=True)
files.download('receipts.csv')
```
